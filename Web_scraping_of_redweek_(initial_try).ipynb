```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
source = requests.get('https://www.redweek.com/resort/P1550-marriotts-newport-coast-villas/timeshare-rentals').text
```

```python
soup = BeautifulSoup(source, 'lxml')
```

```python
tabs = []
table = soup.find("table", class_='table-listing table-listing-rentals table-listing-desktop')
for tr in table.find_all("tr")[1:200]:
  tabs.append({
            'Listing status': tr.find_all('td')[0].text,
            'Dates': tr.find_all('td')[1].text,
            'Nights': tr.find_all('td')[2].text,
            'Price': tr.find_all('td')[3].text,
            'Unit view': tr.find_all('td')[4].text,
            'Beds/Baths': tr.find_all('td')[5].text,
            'Sleeps': tr.find_all('td')[6].text,
            'Listing action': tr.find_all('td')[7].text
            })
df = pd.DataFrame(tabs)
df = df.replace('\n','', regex=True)
df.head(50)
```

```python
from datetime import date
today = date.today()

from google.colab import  drive
drive.mount('/drive')
df.to_csv(f'/drive/Shared drives/All Files/Timeshare platform/Python scripts/Marriott Newport Coast Villas/{today}.csv')
```

```python
df = df[(df['Listing action'] != 'Rented!')]
df.head()
```

```python
df['Listing status'].value_counts()
```

```python
df['Listing status'] = df['Listing status'].replace({'':'Unprotected','RedWeek Verified':'Unprotected','Availability varies':'Unprotected','Verified & Protected':'Protected'})
```

```python
df2 = df.groupby(['Dates','Listing status']).Price.agg([('count', 'count'), ('price', ', '.join)])
df2.head(50)
```
