```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
source = requests.get('https://www.redweek.com/resort/P1550-marriotts-newport-coast-villas/timeshare-rentals').text
```

```python
soup = BeautifulSoup(source, 'lxml')
```

```python
tabs = []
table = soup.find("table", class_='table-listing table-listing-rentals table-listing-desktop')
for tr in table.find_all("tr")[1:200]:
  tabs.append({
            'Listing status': tr.find_all('td')[0].text,
            'Dates': tr.find_all('td')[1].text,
            'Nights': tr.find_all('td')[2].text,
            'Price': tr.find_all('td')[3].text,
            'Unit view': tr.find_all('td')[4].text,
            'Beds/Baths': tr.find_all('td')[5].text,
            'Sleeps': tr.find_all('td')[6].text,
            'Listing action': tr.find_all('td')[7].text
            })
df = pd.DataFrame(tabs)
df = df.replace('\n','', regex=True)
df.head(50)
```

```python
df['Listing action'].value_counts()
```

```python
df = df[(df['Listing action'] != 'Rented!')]
df.head()
```

```python
df['Listing status'].value_counts()
```

```python
df['Listing status'] = df['Listing status'].replace({'':'Unprotected','RedWeek Verified':'Unprotected','Availability varies':'Unprotected','Verified & Protected':'Protected'})
```

```python
df = df.replace(' New!','', regex=True)
```

```python
df = df[['Listing status','Dates','Price']]
```

```python
!pip install xlsxwriter
import xlsxwriter

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter(f'/drive/Shared drives/All Files/Timeshare platform/Python scripts/Marriott trial/final_{today}.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
df.to_excel(writer, sheet_name='Sheet1')

# Get the xlsxwriter workbook and worksheet objects.
workbook  = writer.book
worksheet = writer.sheets['Sheet1']

#Color format
red_format = workbook.add_format({'bg_color': '#FFC7CE',
                               'font_color': '#9C0006'})
green_format = workbook.add_format({'bg_color': '#C6EFCE',
                               'font_color': '#006100'})
black_format = workbook.add_format({'bg_color': '#000000',
                               'font_color': '#000000'})
white_format = workbook.add_format({'bg_color': '#FFFFFF',
                               'font_color': '#FFFFFF'})

# Apply a conditional format to the cell range.
worksheet.conditional_format('E2:E1000', {'type':     'text',
                                       'criteria': 'containing',
                                       'value':    'New Listing',
                                       'format':   green_format})
worksheet.conditional_format('E2:E1000', {'type':     'text',
                                       'criteria': 'containing',
                                       'value':    'Rented',
                                       'format':   red_format})
worksheet.conditional_format('F2:E1000', {'type':     'cell',
                                       'criteria': '==',
                                       'value':    '1',
                                       'format':   black_format})
worksheet.conditional_format('F2:E1000', {'type':     'cell',
                                       'criteria': '==',
                                       'value':    '2',
                                       'format':   white_format})

# Set Auto column width
def get_col_widths(dataframe):
    # First we find the maximum length of the index column   
    idx_max = max([len(str(s)) for s in dataframe.index.values] + [len(str(dataframe.index.name))])
    # Then, we concatenate this to the max of the lengths of column name and its values for each column, left to right
    return [idx_max] + [max([len(str(s)) for s in dataframe[col].values] + [len(col)]) for col in dataframe.columns]

for i, width in enumerate(get_col_widths(df)):
    worksheet.set_column(i, i, width)

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
